```python
emojis_index = [1318, 7305, 12241, 14425, 15720, 17016, 20017, 21229, 24189, 37377, 43236]
```

```python
df.drop(emojis_index, inplace = True)
```

```python
df.reset_index(inplace=True)
```

```python
rows = list(df.index)
```

```python
reading_score = [readability.getmeasures(df['reviewText'][i], lang = 'en')['readability grades']['FleschReadingEase'] for i in rows]
```

```python
print(reading_score)
```

```python
reading_score_textstat = [textstat.flesch_reading_ease(df['reviewText'][i]) for i in rows]
```

```python
print(reading_score_textstat)
```

```python
reading_score_own = [readability.FleschReadingEase(syllables.estimate(df['reviewText'][i]), len(df['reviewText'][i].split()), len(sent_tokenize(df['reviewText'][i]))) for i in rows]
```

```python
print(reading_score_own)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity([reading_score], [reading_score_textstat])
```

```python
cosine_similarity([reading_score], [reading_score_own])
```

```python
cosine_similarity([reading_score_own], [reading_score_textstat])
```

```python
df['reading_score'] = reading_score
df['reading_score_textstat'] = reading_score_textstat
df['reading_score_own'] = reading_score_own
```
