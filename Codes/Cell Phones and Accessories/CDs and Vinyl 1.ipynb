```python
import os
import json
import gzip
import pandas as pd
import numpy as np
```

```python
with open(r"C:\Users\surya\Desktop\Jupyter Python Programs\Amazon Review NLP\reviews_CDs_and_Vinyl_5.json\CDs_and_Vinyl_5.json") as f:
    all_data = [json.loads(line) for line in f]
```

```python
main_df = pd.DataFrame.from_dict(all_data)
```

```python
for i in main_df.columns:
    main_df[i] = main_df[i].astype(str)
```

```python
main_df[main_df.duplicated(keep=False)]
```

```python
with open(r"C:\Users\surya\Desktop\Jupyter Python Programs\Amazon Review NLP\meta_CDs_and_Vinyl.json\meta_CDs_and_Vinyl.json") as f:
    meta_data = [json.loads(line) for line in f]
```

```python
meta_df = pd.DataFrame.from_dict(meta_data)
```

```python
meta_df
```

```python
for i in meta_df.columns:
    meta_df[i] = meta_df[i].astype(str)
```

```python
duplicate = meta_df[meta_df.duplicated(keep=False)]
duplicate.sort_values(by = 'description')
```

```python
clean_meta_df = meta_df.drop_duplicates()
clean_meta_df
```

```python
from functools import reduce

common_asin = list(reduce(set.intersection, map(set, [clean_meta_df.asin, main_df.asin])))
common_asin
```

```python
len(common_asin)
```

```python
combined_df = pd.merge(main_df, clean_meta_df, on='asin')
```

```python
combined_df[combined_df.duplicated(keep=False)]
```

```python
df = combined_df.drop_duplicates() # redundant but ok
```

```python
df.reset_index(inplace=True)
```

```python
df.sort_values('asin')
```
