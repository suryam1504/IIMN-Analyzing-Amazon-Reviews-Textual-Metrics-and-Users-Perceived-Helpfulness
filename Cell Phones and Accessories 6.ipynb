```python
words = []
unique_words = []
for i in [0,1,2]:
    review = df.iloc[i]['reviewText']
    print(review)
    tokenized_words = review.split()
    print(tokenized_words)
    # no of words
    no_of_words = len(tokenized_words)
    print(no_of_words)
    words.append(no_of_words)
    # no of unique words
    no_of_unique_words = len(set(tokenized_words))
    print(no_of_unique_words)
    unique_words.append(no_of_unique_words)
```

```python
words, unique_words
```

```python
words = []
unique_words = []
for i in list(df.index):
    review = df.iloc[i]['reviewText']

    tokenized_words = review.split()
    
    # no of words
    no_of_words = len(tokenized_words) 
    words.append(no_of_words)
    
    # no of unique words
    no_of_unique_words = len(set(tokenized_words))
    unique_words.append(no_of_unique_words)
```

```python
df['no_of_words'] = words
df['no_of_unique_words'] = unique_words
```
