```python
import json
import pandas as pd
import numpy as np
import pickle
```

### Importing main review data which is in json, making it dataframe, removing duplicates, and pickling it

```python
with open(r"C:\Users\surya\Desktop\Jupyter Python Programs\Amazon Review NLP\Cell_Phones_and_Accessories_5.json\Cell_Phones_and_Accessories_5.json") as f:
    all_data = [json.loads(line) for line in f]
```

```python
all_data_df = pd.DataFrame.from_dict(all_data)
```

```python
all_data_df.head()
```

```python
for i in all_data_df.columns:
    all_data_df[i] = all_data_df[i].astype(str)
```

```python
clean_all_data_df = all_data_df.drop_duplicates()
clean_all_data_df
```

```python
pickle_out = open(r"C:\Users\surya\Desktop\Jupyter Python Programs\Amazon Review NLP\clean_all_data_df.pickle", 'wb')
pickle.dump(clean_all_data_df, pickle_out)
pickle_out.close()
```

```python
pickle_in = open(r"C:\Users\surya\Desktop\Jupyter Python Programs\Amazon Review NLP\clean_all_data_df.pickle", 'rb')
clean_all_data_df = pickle.load(pickle_in)
clean_all_data_df
```
